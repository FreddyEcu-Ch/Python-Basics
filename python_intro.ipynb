### 2.2) Loops (for, while)

Loops are important in Python or in any other programming language as they help you to execute a **block of code repeatedly**. #### 2.2.1. for Loop

- A for loop is a programming concept that, when it's implemented, executes a piece of code over and over again "for" a certain number of times, based on a sequence. This "number of times" is determined by a sequence or an ordered list of things. 

- The syntax of this compound statement is shown below:


![Picture Title](resources/images/for_sintaxis.png)

```python
for crudo in API:
    if crudo <= 10:
        print('El crudo es extra pesado')
    elif (crudo >= 11 and crudo <= 21):
        print("El crudo es pesado")
    elif (crudo >= 22 and crudo <= 29):
        print("El crudo es mediano")
    elif crudo >= 30:
        print("El crudo es liviano")
```

**Índice de Productividad**

Sí $P_{r}$ >= $P_{b}$:\
$$J=\frac{Q_{o}}{P_{r} - Pwf}$$ \
De lo contrario: \
$$J = \frac{Qo_{test}}{(P_{r} - P_{b}) + \frac{P_{b}}{1.8}+(1-0.2\frac{Pwf_{test}}{P_{b}}-0.8\frac{Pwf_{test}}{P_{b}})}$$ \
**Where:** \
*$J$:* Productivity Index (bpd/psi)\
*$Q_{o}$:* Oil Flow Rate (bpd)\
*$P_{r}$:* Reservoir Pressure (psia)\
*$Pwf$:* Pressure Well Flowing (psia)\
*$P_{b}$:* Bubble point pressure (psia)\
*$Pwf_{test}$:* Test Pressure (psia)\
*$Qo_{test}$:* Oil Flow Rate of test (bpd)

```python
# Índice de Productividad
def j(q_test, pwf_test, pr, pb):
    if pwf_test >= pb: #Yacimiento subsaturado
        J = q_test / (pr - pwf_test)
    else: #Yacimiento saturado
        J = q_test / ((pr - pb) + (pb / 1.8) * (1 - 0.2 * (pwf_test / pb) - 0.8 * (pwf_test / pb)**2))
    return J
```

```python
import math
# Presión de Fondo Fluyente en el Medio Poroso
def presion(pi, qo, uo, bo, ko, h, re, rw, s, flujo='continuo'):
    if flujo == 'continuo':
        pwf = pi - (141.2 * qo * bo * uo / ko * h) * (math.log(re / rw) + s)
    elif flujo == 'semicontinuo':
        pwf = pi - (141.2 * qo * bo * uo / ko * h) * (math.log(re / rw) + s)
    return pwf
```

```python
# Resistividad del Agua - Ecuación de Archie
def resistividad_agua(ro, phi, a, m):
    rw = ro * phi ** m / a
    return rw
```

The Volve field start producing in January of 2008 to June 2016. Let's check for how many days the field produced!

```python
import datetime
start_production = datetime.datetime(year=2008, month=1, day=1)
end_production = datetime.datetime(year=2016, month=6, day=1)
prod_days = end_production - start_production
print(prod_days)
```
